```python
import random
```

```python
moods = ["happy", "sad", "calm", "angry"]
budgets = ["high", "medium", "low", "random"]
cities = ["jakarta", "bandung", "semarang", "jogja", "surabaya","random"]
distances = ["high", "medium", "low", "random"]
```

```python
users = {}
```

```python

n = 40
for i in range(n):
    id = random.randint(0,300)
    preferences = {}
    preferences["mood"] = random.choice(moods)
    preferences["budget"] = random.choice(budgets)
    preferences["city"] = random.choice(cities)
    preferences["distance"] = random.choice(distances)
    users[f"{id}"] =  preferences
print(users)
```

```python
import pandas as pd
```

```python
df_user_dummy = pd.DataFrame.from_dict(users, orient='index').reset_index()
df_user_dummy.columns = ['ID'] + list(df_user_dummy.columns)[1:]
```
